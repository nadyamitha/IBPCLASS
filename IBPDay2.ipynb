```python
from matplotlib import pyplot as plt

# Assuming df is your DataFrame containing the data
data = df['YearsWithCurrManager']

# Plot histogram
plt.hist(data, bins=20)
plt.title('Years With Current Manager')
plt.xlabel('Years')
plt.ylabel('Frequency')

# Calculate the histogram once
hist_data = plt.hist(data, bins=20)

# Add number tags on each bar
for i in range(len(data)):
    plt.text(hist_data[1][i], hist_data[0][i], str(hist_data[0][i]), ha='center', va='bottom')

#plt.show()
```
